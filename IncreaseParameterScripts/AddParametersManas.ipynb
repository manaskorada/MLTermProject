```python
import pandas as pd
import spacy
from google.colab import files

df = pd.read_csv('train.csv')

# Load spaCy model
nlp = spacy.load('en_core_web_sm')

# Define function to count named entities
def count_named_entities(text):
    doc = nlp(str(text))
    return len(doc.ents)

# Apply to the description column
df['desc_named_entity_count'] = df['Description'].apply(count_named_entities)

# Save the updated file
df.to_csv('train_with_ner.csv', index=False)

# Download it
files.download('train_with_ner.csv')
```

# Title-Description Similarity

```python
!pip install -q sentence-transformers
```

```python
import pandas as pd
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
from google.colab import files

# Load the updated CSV
df = pd.read_csv('train.csv')

# Load the model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Function to compute cosine similarity between title and description
def compute_similarity(title, desc):
    title = str(title)
    desc = str(desc)
    embeddings = model.encode([title, desc])
    return cosine_similarity([embeddings[0]], [embeddings[1]])[0][0]

# Apply to each row
df['title_desc_similarity'] = df.apply(lambda row: compute_similarity(row['Title'], row['Description']), axis=1)

# Save the final version
df.to_csv('train_final.csv', index=False)

# Download it
files.download('train_final.csv')
```
