# Word count Scripts - 6 features
1. Title average word length
2. Description average word length
3. Title word count
4. Description word count
5. Title char count
6. Description char count

```python
"""
  Takes a string and returns the average word length.
  data: string
  return: float
"""
def AvgWordLength(data: str) -> float:
  avg = 0

  words = data.split()
  for word in words:
    avg += len(word)
  avg /= len(words)

  return avg

"""
  Tkes a string and returns the total word count.
  data: string
  return: int
"""
def wordCount(data: str) -> int:
  return len(data.split())

"""
  Takes a string and returns the total character count.
  data: string
  return: int
"""
def charCount(data: str) -> int:
  return len(data)
```

```python
import pandas as pd


test_data = pd.read_csv("test.csv")
test_data.head()
```

```python
train_data = pd.read_csv("train.csv")
train_data.head()
```

```python
"""
  Applies all of the scripts to the title at once.
  Returns reference to the dataframe
"""
def applyScriptsTitle(data: pd.DataFrame) -> pd.DataFrame:
  data['AvgTitleWordLength'] = data['Title'].apply(AvgWordLength)
  data['TitleWordCount'] = data['Title'].apply(wordCount)
  data['TitleCharCount'] = data['Title'].apply(charCount)
  return data

"""
  Applies all of the scripts to the description.
  Returns reference to the dataframe
"""
def applyScriptsDesc(data: pd.DataFrame) -> pd.DataFrame:
  data['DescAvgWordLength'] = data['Description'].apply(AvgWordLength)
  data['DescWordCount'] = data['Description'].apply(wordCount)
  data['DescCharCount'] = data['Description'].apply(charCount)
  return data



applyScriptsTitle(test_data)
applyScriptsTitle(train_data)

applyScriptsDesc(test_data)
applyScriptsDesc(train_data)


test_data.head()
```

# Sentiment Analysis - 8 features
1. Negative sentiment score
2. Neutral sentiment score
3. positive sentiment score
4. Compound sentiment score

x2 for the Title and the Description

For sentiment analysis, we will use the Nltk library and the VADER_lexicon dictionary. (Valence Aware Dictionary and sEntiment Reasoner)

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import nltk # Natural language toolkit
nltk.download('vader_lexicon')
```

```python
"""
  Given a string, will return the sentiment analysis of the string.
  title: string
  return: set {'neg': 0.0, 'neu': 0.468, 'pos': 0.532, 'compound': 0.6369}
"""
def sentimentAnalysis(title: str) -> set:
  sid = SentimentIntensityAnalyzer().polarity_scores(title)
  return sid

"""
  Given the title column of a dataframe, will apply the sentiment analysis to the title.
  data: dataframe
  return: dataframe
"""
def applySentimentTitle(data: pd.DataFrame) -> pd.DataFrame:
  data['Sentiment'] = data['Title'].apply(sentimentAnalysis)
  return data

"""
  Given the description column of a dataframe, will apply the sentiment analysis to the description.
  data: dataframe
  return: dataframe
"""
def applySentimentDesc(data: pd.DataFrame) -> pd.DataFrame:
  data['DescSentiment'] = data['Description'].apply(sentimentAnalysis)
  return data
```

```python
applySentimentTitle(test_data)
applySentimentTitle(train_data)

applySentimentDesc(test_data)
applySentimentDesc(train_data)

test_data.head()
```
