Here we tokenize each title and description and preprocess the word

```python
titles = []
for sent in trainingData['Title']:
  sent_token = sent_tokenize(sent)
  for word in sent_token:
    titles.append(simple_preprocess(word))

descriptions = []
for sent in trainingData['Description']:
  sent_token = sent_tokenize(sent)
  for word in sent_token:
    descriptions.append(simple_preprocess(word))


print(len(titles))
print(len(descriptions))
```

```python
testTitles = []
for sent in testingData['Title']:
  sent_token = sent_tokenize(sent)
  for word in sent_token:
    testTitles.append(simple_preprocess(word))

testDescriptions = []
for sent in testingData['Description']:
  sent_token = sent_tokenize(sent)
  for word in sent_token:
    testDescriptions.append(simple_preprocess(word))


print(testTitles[2])
print(testDescriptions[2])
```

# Creating the feature vectors
Now that we have our word vectors, we need to properly format our features. We want an input shape of (21,), so that means we need to convert all of our features into float64 from int64

```python
trainingData.info()
```

```python
print(trainingData['DescNamedEntityCount'].dtype)

trainingData['DescNamedEntityCount'] = trainingData['DescNamedEntityCount'].astype(np.float64)

print(trainingData['DescNamedEntityCount'].dtype)
```

Now we do this for the testing data, and for all the other non-float64 data types

```python
testingData['DescNamedEntityCount'] = testingData['DescNamedEntityCount'].astype(np.float64)
trainingData['TitleWordCount'] = trainingData['TitleWordCount'].astype(np.float64)
testingData['TitleWordCount'] = testingData['TitleWordCount'].astype(np.float64)
trainingData['TitleCharCount'] = trainingData['TitleCharCount'].astype(np.float64)
testingData['TitleCharCount'] = testingData['TitleCharCount'].astype(np.float64)
trainingData['DescWordCount'] = trainingData['DescWordCount'].astype(np.float64)
testingData['DescWordCount'] = testingData['DescWordCount'].astype(np.float64)
trainingData['DescCharCount'] = trainingData['DescCharCount'].astype(np.float64)
testingData['DescCharCount'] = testingData['DescCharCount'].astype(np.float64)
trainingData['UppercaseWordCount'] = trainingData['UppercaseWordCount'].astype(np.float64)
testingData['UppercaseWordCount'] = testingData['UppercaseWordCount'].astype(np.float64)

trainingData.info()
```

# The Model

```python
from keras.models import Sequential, Model
from keras.layers import Dense, Embedding, LSTM, SpatialDropout1D, Input, Concatenate, Dropout
from keras.callbacks import EarlyStopping
```

```python
# Title and description fed through LSTM
text_input = Input(shape=(80,), name='text_input')
embedding = Embedding(39155, 128, input_length=80)(text_input)  # Input as the number of words (vocabulary size)
lstm_out = LSTM(100, dropout=0.2, recurrent_dropout=0.2)(embedding)

# Other features in dense layer
feature_input = Input(shape=(21,), name='feature_input')
feature_dense = Dense(64, activation = 'relu')(feature_input)

# Merge both layers
merged = Concatenate()([lstm_out, feature_dense])
merge = Dense(128, activation = 'relu')(merged)
merged = Dropout(0.3)(merged)
output = Dense(4, activation='softmax')(merged)


model1 = Model(inputs = [text_input, feature_input], outputs=output) #Softmax for multiclass
model1.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
accuracy = model1.evaluate(
    {'text_input': testArticles, 'feature_input':testingFeatures},
    testingClasses
    )
print(accuracy[0],accuracy[1])
```
